```python
def new_locality_name(loc_name):
    words = loc_name.split(' ')
    new_name = []
    for word in words:
        if (
            '–ø–æ—Å–µ' not in word and
            '–ø–æ—Å—ë' not in word and
            '–≥–æ—Ä–æ–¥' not in word and
            '–¥–µ—Ä–µ–≤' not in word and
            '—Å–µ–ª–æ' not in word and
            '—Å–∞–¥–æ–≤' not in word and
            '—Ç–æ–≤–∞—Ä–∏—â–µ—Å—Ç–≤' not in word and
            '—Å–µ–ª–æ' not in word and
            '–ø—Ä–∏' not in word and
            '–∂–µ–ª–µ–∑–Ω–æ–¥–æ—Ä' not in word and
            '—Å—Ç–∞–Ω—Ü' not in word and
            '—Ç–∏–ø' not in word):
 
            new_name.append(word)
 
    return ' '.join(new_name)
```

<div class="alert alert-block alert-warning">
<b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞‚ö†Ô∏è:</b> –õ–∏—à–Ω—è—è —è—á–µ–π–∫–∞?

```python
data['TOP_10_cities'] = data['only_locality_name'].apply(new_locality_name)
```
